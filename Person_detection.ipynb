```
!wget 'https://github.com/OlafenwaMoses/ImageAI/releases/download/1.0/resnet50_coco_best_v2.0.1.h5' 
```

```
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath('/content/resnet50_coco_best_v2.0.1.h5')
detector.loadModel()
```

```
out = cv2.VideoWriter('outpy.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 1280, (480,360))

vid = cv2.VideoCapture('/content/drive/My Drive/videoplayback.mp4') 

Dataset = []
```

```
i = 0
while(vid.isOpened()):
    
    ret, frame = vid.read() 
    
    if ret == True:
        
        t = datetime.now() 
        in_img = "/content/supply/"+  "imagenew"+str(i)+".jpg"
        cv2.imwrite(in_img,frame)

        out_img = "/content/store/"+  "imagenew"+str(i)+".jpg"
        detections = detector.detectObjectsFromImage(input_image= in_img , output_image_path= out_img)
        
        mod_img = cv2.imread(out_img)

        out.write(mod_img)
        
        i+=1
        #person =[] 
        count = 0 
        print(i)
        for eachObject in detections:
            if eachObject["name"] == 'person':
                #person.append(eachObject["percentage_probability"])
                count+=1
        
        Dataset.append([count,t])

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    else:
        
        break
    
vid.release()
out.release()

cv2.destroyAllWindows()
```
